```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
```

```python
csv1_path = r"Student_Performance_1.csv"
df1 = pd.read_csv(csv1_path)
```

```python
csv2_path = r"Student_Performance_2.csv"
df2 = pd.read_csv(csv2_path)
```

```python
df = pd.concat([df1, df2])
```

```python
df
```

```python
print("\n First 5 rows\n")
df.head()
```

```python
print("\n Last 5 rows of dataset\n")
df.tail()
```

```python
print("\nData Info:")
df.info()
```

```python
print("Description of data:\n")
df.describe()
```

```python
print("Identifying null values in dataset\n")
df.isnull()
```

```python
print("\n Total number of missing values in dataset per column\n")
df.isnull().sum()
```

```python
print("\n Handling missing values in dataset\n")
# Add Message column and fill with 'welcome'
df['Message'] = 'welcome'
df.isnull().sum()
```

```python
print("Identifying duplicates in dataset:")
df.duplicated().sum()
```

```python
print("\n Removing duplicates in the dataset\n")
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
print("Normalization of data :\n")
scaler = MinMaxScaler()
numeric_cols = df.select_dtypes(include=["float64", "int64"]).columns
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])
print("\n Dataset after normalization: \n")
df
```

```python
df = pd.get_dummies(df, columns=['Message'])
```

```python
df
```

```python
df.to_csv('student_performance_processed.csv', index=False)
print("\nCleaned data saved to 'student_performance_processed.csv'")
```

```python
# Create histogram visualization (similar to the original file)
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
df[numeric_cols].hist(figsize=(10, 6))
plt.suptitle('Histograms of Scaled Student Performance Features')
plt.tight_layout()
plt.show()
```
